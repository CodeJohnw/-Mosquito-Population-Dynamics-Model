```python
####
import numpy as np
import matplotlib.pyplot as plt

# Parameter settings
N0 = 1000         # Initial adult mosquito population
r = 0.1          # Egg laying rate per day
mu = 0.05        # Death rate per day
K = 5000         # Environmental carrying capacity
tau = 10         # Delay time (days)
T = 200          # Total simulation time (days)
dt = 1           # Time step (days)

# Initialize time and population arrays
time = np.arange(0, T+dt, dt)
N = np.zeros(len(time))

# Initial condition
N[0] = N0

# Initialize the effect of delay (using the initial population)
for i in range(1, tau+1):
    N[i] = N0

# Solve the differential equation using Euler's method
for i in range(tau, len(time)-1):
    dNdt = r*N[i]*(1 - N[i]/K) - mu*N[i] + r*N[i-tau]*(1 - N[i-tau]/K)
    N[i+1] = N[i] + dNdt*dt

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(time, N, label='Mosquito Population')
plt.title('Mosquito Population Dynamics Model Considering Carrying Capacity')
plt.xlabel('Time (days)')
plt.ylabel('Number of Mosquitoes')
plt.legend()
plt.grid(True)
plt.show()
```

```python
##
import numpy as np
import matplotlib.pyplot as plt

# 
r = 0.5  # 
K = 100000  # 
mu = 0.1  # 
mu_c = 0.09  # 
dt = 0.01  # 
T = 100  # 
N0 = 1000  # 

# 
t = np.arange(0, T, dt)
N = np.zeros(len(t))
N[0] = N0

# Euler
for i in range(1, len(t)):
    N[i] = N[i-1] + dt * (r * N[i-1] * (1 - N[i-1] / K) - (mu + mu_c) * N[i-1])

# 
plt.figure(figsize=(10, 6))
plt.plot(t, N, label='Mosquito Population')
plt.xlabel('Time(day)')
plt.ylabel('Number of Mosquitoes')
plt.title('Mosquito Population Dynamics Model Considering Carrying Capacity and Control Measures')
plt.legend()
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# 
N0 = 1000  # 
r = 0.2  # 
mu = 0.01  # 
mu_c = 0.0  # 
K = 5000  # 
tau = 14  # 
T = 100  # 
dt = 0.1  # 

# 
steps = int(T / dt)  # 
N = np.zeros(steps)  # 
N[0] = N0  # 

# 
for t in range(1, steps):
    t_delay = int(t - tau / dt)  # 
    if t_delay < 0:
        N_delayed = N0
    else:
        N_delayed = N[t_delay]
    
    dNdt = r * N_delayed * (1 - N_delayed / K) - (mu + mu_c) * N[t - 1]  # 
    N[t] = N[t - 1] + dNdt * dt  # 

# 
plt.figure(figsize=(10, 6))
plt.plot(np.arange(0, T, dt), N, label='Model 3: With control measures')
plt.xlabel('Time (days)')
plt.ylabel('Mosquito population')
plt.title('Mosquito Population Dynamics')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 
r = 0.2  # 
K = 5000  # 
mu = 0.02  # 
mu_c = 0.0  # 
tau = 20  # 
T = 200  # 
dt = 1  # 

# 
N = np.zeros(T)  # 
N[0] = 500  # 

# Euler
for t in range(0, T-1):
    if t < tau:
        # tau
        N_delayed = N[0]
    else:
        # tau
        N_delayed = N[t - tau]
    
    dNdt = r * N_delayed * (1 - N[t] / K) - (mu + mu_c) * N[t]
    N[t + 1] = N[t] + dt * dNdt

# 
plt.figure(figsize=(10, 6))
plt.plot(N, label='N(t)')
plt.xlabel('Time (days)')
plt.ylabel('Population')
plt.title('Mosquito Population Dynamics with Initial Egg-Laying')
plt.legend()
plt.show()
```
