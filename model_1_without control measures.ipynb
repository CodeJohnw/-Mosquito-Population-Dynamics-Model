```python
####考虑环境容量、未考虑控制措施、考虑时间延迟
import numpy as np
import matplotlib.pyplot as plt

# Parameter settings
N0 = 1000         # Initial adult mosquito population
r = 0.1          # Egg laying rate per day
mu = 0.05        # Death rate per day
K = 5000         # Environmental carrying capacity
tau = 10         # Delay time (days)
T = 200          # Total simulation time (days)
dt = 1           # Time step (days)

# Initialize time and population arrays
time = np.arange(0, T+dt, dt)
N = np.zeros(len(time))

# Initial condition
N[0] = N0

# Initialize the effect of delay (using the initial population)
for i in range(1, tau+1):
    N[i] = N0

# Solve the differential equation using Euler's method
for i in range(tau, len(time)-1):
    dNdt = r*N[i]*(1 - N[i]/K) - mu*N[i] + r*N[i-tau]*(1 - N[i-tau]/K)
    N[i+1] = N[i] + dNdt*dt

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(time, N, label='Mosquito Population')
plt.title('Mosquito Population Dynamics Model Considering Carrying Capacity')
plt.xlabel('Time (days)')
plt.ylabel('Number of Mosquitoes')
plt.legend()
plt.grid(True)
plt.show()
```

```python
##考虑环境容量、考虑控制措施、未考虑时间延迟
import numpy as np
import matplotlib.pyplot as plt

# 参数设定
r = 0.5  # 自然生长率
K = 100000  # 环境容量
mu = 0.1  # 自然死亡率
mu_c = 0.09  # 控制措施导致的额外死亡率
dt = 0.01  # 时间步长
T = 100  # 总时间
N0 = 1000  # 初始种群数量

# 时间序列
t = np.arange(0, T, dt)
N = np.zeros(len(t))
N[0] = N0

# Euler法数值解
for i in range(1, len(t)):
    N[i] = N[i-1] + dt * (r * N[i-1] * (1 - N[i-1] / K) - (mu + mu_c) * N[i-1])

# 绘图
plt.figure(figsize=(10, 6))
plt.plot(t, N, label='Mosquito Population')
plt.xlabel('Time(day)')
plt.ylabel('Number of Mosquitoes')
plt.title('Mosquito Population Dynamics Model Considering Carrying Capacity and Control Measures')
plt.legend()
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# 参数设定
N0 = 1000  # 初始成年蚊子数量
r = 0.2  # 繁殖率
mu = 0.01  # 自然死亡率
mu_c = 0.0  # 控制措施导致的额外死亡率
K = 5000  # 环境容量
tau = 14  # 成长延迟时间（天）
T = 100  # 总模拟时间（天）
dt = 0.1  # 时间步长（天）

# 初始化
steps = int(T / dt)  # 计算总步数
N = np.zeros(steps)  # 创建数组存储每一步的蚊子数量
N[0] = N0  # 初始值

# 数值求解循环
for t in range(1, steps):
    t_delay = int(t - tau / dt)  # 计算延迟对应的步数
    if t_delay < 0:
        N_delayed = N0
    else:
        N_delayed = N[t_delay]
    
    dNdt = r * N_delayed * (1 - N_delayed / K) - (mu + mu_c) * N[t - 1]  # 模型方程
    N[t] = N[t - 1] + dNdt * dt  # 更新蚊子数量

# 结果绘图
plt.figure(figsize=(10, 6))
plt.plot(np.arange(0, T, dt), N, label='Model 3: With control measures')
plt.xlabel('Time (days)')
plt.ylabel('Mosquito population')
plt.title('Mosquito Population Dynamics')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 参数设置
N0 = 100         # 初始成年蚊子数量
r = 0.1          # 产卵率
mu = 0.05        # 死亡率
K = 1000         # 环境容量
tau = 10         # 延迟时间（天）
T = 100          # 总模拟时间（天）
dt = 1           # 时间步长（天）

# 时间和种群数量数组初始化
time = np.arange(0, T+dt, dt)
N = np.zeros(len(time))

# 初始条件
N[0] = N0

# 延迟影响的初始化（使用初始数量）
for i in range(1, tau+1):
    N[i] = N0

# Euler方法求解微分方程
for i in range(tau, len(time)-1):
    dNdt = r*N[i]*(1 - N[i]/K) - mu*N[i] + r*N[i-tau]*(1 - N[i-tau]/K)
    N[i+1] = N[i] + dNdt*dt

# 绘图
plt.figure(figsize=(10, 6))
plt.plot(time, N, label='蚊子种群数量')
plt.title('考虑环境容量限制的蚊子种群动态模型')
plt.xlabel('时间 (天)')
plt.ylabel('蚊子数量')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 参数设置
r = 0.2  # 日产卵率
K = 5000  # 环境容量
mu = 0.02  # 自然死亡率
mu_c = 0.0  # 控制措施导致的额外死亡率
tau = 20  # 时间延迟（天）
T = 200  # 总模拟时间（天）
dt = 1  # 时间步长（天）

# 初始化
N = np.zeros(T)  # 成年蚊子的数量
N[0] = 500  # 初始成年蚊子数量

# Euler方法求解微分方程
for t in range(0, T-1):
    if t < tau:
        # 对于初始的tau天，使用初始成年蚊子数量作为延迟期的近似
        N_delayed = N[0]
    else:
        # 对于tau天之后，使用真实的历史数据
        N_delayed = N[t - tau]
    
    dNdt = r * N_delayed * (1 - N[t] / K) - (mu + mu_c) * N[t]
    N[t + 1] = N[t] + dt * dNdt

# 绘制结果
plt.figure(figsize=(10, 6))
plt.plot(N, label='N(t)')
plt.xlabel('Time (days)')
plt.ylabel('Population')
plt.title('Mosquito Population Dynamics with Initial Egg-Laying')
plt.legend()
plt.show()
```
